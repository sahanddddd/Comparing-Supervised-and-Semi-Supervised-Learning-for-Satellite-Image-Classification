```python
# Define constants
IMG_SIZE = (224, 224)  # Input size for EfficientNet-Lite
BATCH_SIZE = 32
NUM_CLASSES = 10

UNLABELED_DATA_DIR = dirc + '/SplitDataset/unlabeled_flat'
TRAIN_DIR = dirc + '/SplitDataset/labeled'
TEST_DIR = dirc + '/SplitDataset/test'
EFFICIENTNET_MODEL_PATH = dirc + '/models/eurosat_efficientnetlite_model.h5'
```

```python
from tensorflow.keras.models import load_model
```

```python
class EfficientNetLiteWrapper(tf.keras.layers.Layer):
    def __init__(self, url, **kwargs):
        super(EfficientNetLiteWrapper, self).__init__(**kwargs)
        self.base_model = hub.KerasLayer(url, trainable=False)

    def call(self, inputs, training=False):
        return self.base_model(inputs, training=training)
```

```python
model = load_model(
    EFFICIENTNET_MODEL_PATH,
    custom_objects={
        'KerasLayer': hub.KerasLayer,
        'EfficientNetLiteWrapper': EfficientNetLiteWrapper
    },
    compile=False  # Prevents issues with optimizer states
)

print("Model Loaded")
```

```python
# Freeze the base model for inference mode
for layer in model.layers:
    layer.trainable = False
```

```python
# Create a function to generate pseudo-labels
def generate_pseudo_labels(model, unlabeled_data_dir, img_size, batch_size):
    # Get all image file paths
    unlabeled_filepaths = [os.path.join(unlabeled_data_dir, fname) for fname in os.listdir(unlabeled_data_dir)]
    
    # Create a generator for unlabeled data
    unlabeled_datagen = ImageDataGenerator(rescale=1./255)
    unlabeled_generator = unlabeled_datagen.flow_from_directory(
        directory=os.path.dirname(unlabeled_data_dir),
        classes=[os.path.basename(unlabeled_data_dir)],
        target_size=img_size,
        batch_size=batch_size,
        class_mode=None,  # No labels
        shuffle=False
    )
    
    # Predict pseudo-labels
    pseudo_labels = model.predict(unlabeled_generator, steps=len(unlabeled_generator), verbose=1)
    pseudo_labels = np.argmax(pseudo_labels, axis=1)  # Convert probabilities to class indices
    
    return unlabeled_filepaths, pseudo_labels
```

```python
# Generate pseudo-labels for unlabeled data
unlabeled_filepaths, pseudo_labels = generate_pseudo_labels(model, UNLABELED_DATA_DIR, IMG_SIZE, BATCH_SIZE)
```

```python
# Combine labeled and pseudo-labeled data
def combine_labeled_and_pseudo_labeled(train_dir, unlabeled_filepaths, pseudo_labels, img_size, batch_size):
    # Create generators for labeled data
    train_datagen = ImageDataGenerator(rescale=1./255)
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True
    )
    
    # Create a DataFrame for pseudo-labeled data
    import pandas as pd
    pseudo_df = pd.DataFrame({
        'filename': unlabeled_filepaths,
        'class': pseudo_labels
    })
    
    # Map class indices to class names
    class_indices = {v: k for k, v in train_generator.class_indices.items()}
    pseudo_df['class'] = pseudo_df['class'].map(class_indices)
    
    # Save pseudo-labeled data to a temporary directory
    temp_dir = '/kaggle/working/pseudo_labeled_data'
    os.makedirs(temp_dir, exist_ok=True)
    for _, row in pseudo_df.iterrows():
        class_name = row['class']
        os.makedirs(os.path.join(temp_dir, class_name), exist_ok=True)
        os.symlink(row['filename'], os.path.join(temp_dir, class_name, os.path.basename(row['filename'])))
    
    # Create a generator for combined data
    combined_datagen = ImageDataGenerator(rescale=1./255)
    combined_generator = combined_datagen.flow_from_directory(
        temp_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True
    )
    
    return combined_generator
```

```python
# Combine labeled and pseudo-labeled data
combined_generator = combine_labeled_and_pseudo_labeled(TRAIN_DIR, unlabeled_filepaths, pseudo_labels, IMG_SIZE, BATCH_SIZE)
```

```python
# Implement FixMatch
def fixmatch_weak_augmentation(img_size):
    return ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
```

```python
def fixmatch_strong_augmentation(img_size):
    return ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.3,
        height_shift_range=0.3,
        shear_range=0.2,
        zoom_range=0.3,
        horizontal_flip=True,
        brightness_range=(0.7, 1.3),
        fill_mode='nearest'
    )
```

```python
# Create weak and strong augmentation generators
weak_aug_gen = fixmatch_weak_augmentation(IMG_SIZE).flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)
```

```python
strong_aug_gen = fixmatch_strong_augmentation(IMG_SIZE).flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)
```

```python
# Train the model with combined data
def train_with_combined_data(model, combined_generator, test_dir, img_size, batch_size, epochs):
    # Create a generator for test data
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )
    
    # Compile the model
    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    # Define callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
    
    # Train the model
    history = model.fit(
        combined_generator,
        steps_per_epoch=combined_generator.samples // batch_size,
        epochs=epochs,
        validation_data=test_generator,
        validation_steps=test_generator.samples // batch_size,
        callbacks=[early_stopping, reduce_lr]
    )
    
    return history
```

```python
# Train the model
history = train_with_combined_data(model, combined_generator, TEST_DIR, IMG_SIZE, BATCH_SIZE, epochs=20)
```

```python
# Fine-tune the model (optional)
for layer in model.layers:
    layer.trainable = True

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

**Below is the code to further fine tune the model**
Fine-tuning is typically done after the initial training phase to further improve the model's performance. Here's why it's important:

**Unfreezing Layers :**
Before fine-tuning, all layers of the pre-trained model were frozen (i.e., their weights were not updated during training).
During fine-tuning, the layers are unfrozen (layer.trainable = True), allowing the model to adapt its pre-trained weights to the new task.

But as shown in the above picture, each epoch will require 4 to 5 hours of training on GPU. Due to a lack of resources and the fact that the baseline efficientnet model was providing us with the highest accuracy, we assume that the model after fine-tuning will outperform all three baseline models.

```python
# Create a global test generator
def create_test_generator(test_dir, img_size, batch_size):
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )
    return test_generator

# Create the test generator
test_generator = create_test_generator(TEST_DIR, IMG_SIZE, BATCH_SIZE)
```

```python
fine_tune_history = model.fit(
    combined_generator,
    steps_per_epoch=combined_generator.samples // BATCH_SIZE,
    epochs=10,
    initial_epoch=history.epoch[-1],
    validation_data=test_generator,
    validation_steps=test_generator.samples // BATCH_SIZE,
    callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
               ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)]
)

# Save the final model
model.save('/kaggle/working/final_ssl_model.h5')
```
