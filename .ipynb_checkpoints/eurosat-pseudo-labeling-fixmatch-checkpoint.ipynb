```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
```

```python
# Get class labels
class_labels = list(train_generator.class_indices.keys())

# Compute class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)

# Convert class weights to dictionary
class_weights_dict = dict(enumerate(class_weights))
class_names = list(train_generator.class_indices.keys())
```

```python
# Create generator for unlabeled data
unlabeled_datagen = ImageDataGenerator(rescale=1./255)

unlabeled_generator = unlabeled_datagen.flow_from_directory(
    directory=os.path.dirname(unlabeled_dir),
    classes=[os.path.basename(unlabeled_dir)],
    target_size=img_size,
    batch_size=batch_size,
    class_mode=None,
    shuffle=False
)

# Generate pseudo-labels
print("Generating pseudo-labels...")
steps = np.ceil(unlabeled_generator.samples / batch_size).astype(int)
predictions = model.predict(unlabeled_generator, steps=steps)

# Get confidence scores and pseudo-labels
confidence_scores = np.max(predictions, axis=1)
pseudo_labels = np.argmax(predictions, axis=1)

# Apply confidence threshold
confident_mask = confidence_scores >= confidence_threshold
confident_indices = np.where(confident_mask)[0]

# Create dataframe for pseudo-labeled data
unlabeled_filenames = [os.path.join(unlabeled_dir, fname.split('/')[-1]) for fname in unlabeled_generator.filenames]
pseudo_df = pd.DataFrame({
    'filename': [unlabeled_filenames[i] for i in confident_indices],
    'class': [class_names[pseudo_labels[i]] for i in confident_indices]  # Use list-converted class names
})

# Create dataframe for original training data
original_train_df = pd.DataFrame({
    'filename': [os.path.join(train_dir, f) for f in train_generator.filenames],
    'class': [class_names[c] for c in train_generator.classes]  # Use list-converted class names
})


# Combine datasets
combined_df = pd.concat([original_train_df, pseudo_df], ignore_index=True)

# Create new data generator for combined data
combined_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

combined_generator = combined_datagen.flow_from_dataframe(
    dataframe=combined_df,
    x_col='filename',
    y_col='class',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    classes=list(train_generator.class_indices.keys()),
    shuffle=True
)
```

```python
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential()

    # First Convolutional Layer
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))

    # Second Convolutional Layer
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Third Convolutional Layer
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten the output
    model.add(layers.Flatten())

    # Fully Connected Layer
    model.add(layers.Dense(128, activation='relu'))

    # Output Layer
    model.add(layers.Dense(num_classes, activation='softmax'))

    return model

# Define input shape and number of classes
input_shape = (img_size[0], img_size[1], 3)  # RGB images
num_classes = 10

# Create the model
model = create_cnn_model(input_shape, num_classes)

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
# Compute new class weights
combined_classes = [list(train_generator.class_indices.keys()).index(c) for c in combined_df['class']]
class_weights = compute_class_weight('balanced', classes=np.unique(combined_classes), y=combined_classes)
class_weights_dict = dict(enumerate(class_weights))
```

```python
# Fine-tune the model with pseudo-labeled data
print("Training with pseudo-labeled data...")
epochs_pseudo = 20

history_pseudo = model.fit(
    combined_generator,
    steps_per_epoch=combined_generator.samples // batch_size,
    epochs=epochs_pseudo,
    validation_data=test_generator,
    validation_steps=test_generator.samples // batch_size,
    class_weight=class_weights_dict
)
```

**FixMatch**

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import models, layers
import os
import numpy as np
import pandas as pd
from sklearn.utils.class_weight import compute_class_weight

# Define paths and parameters
train_dir = 'SplitDataset/labeled'
test_dir = 'SplitDataset/test'
unlabeled_dir = 'SplitDataset/unlabeled_flat'
batch_size = 64
img_size = (64, 64)
confidence_threshold = 0.75  
lambda_u = 1.0  # Weight for unsupervised loss
```

```python
# Load pre-trained model or create a new one
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(num_classes, activation='softmax'))
    return model

input_shape = (img_size[0], img_size[1], 3)
num_classes = 10
model = create_cnn_model(input_shape, num_classes)

# Compile the model (add this block)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Data generators for labeled data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Strong augmentation for unlabeled data
strong_aug_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.5, 1.5],
    fill_mode='nearest'
)

# Generators for weak and strong augmentations on unlabeled data
unlabeled_weak_generator = train_datagen.flow_from_directory(
    directory=os.path.dirname(unlabeled_dir),
    classes=[os.path.basename(unlabeled_dir)],
    target_size=img_size,
    batch_size=batch_size,
    class_mode=None,
    shuffle=False,
    seed=42
)

unlabeled_strong_generator = strong_aug_datagen.flow_from_directory(
    directory=os.path.dirname(unlabeled_dir),
    classes=[os.path.basename(unlabeled_dir)],
    target_size=img_size,
    batch_size=batch_size,
    class_mode=None,
    shuffle=False,
    seed=42,
)
```

```python
# Verify labeled data
for x, y in train_generator:
    print("Labeled data:", x.shape, y.shape)
    print("Sample labels:", np.unique(np.argmax(y, axis=1), return_counts=True))
    break

# Verify unlabeled data
for weak_x in unlabeled_weak_generator:
    print("Unlabeled data:", weak_x.shape)
    print("Sample pixel range:", weak_x.min(), weak_x.max())
    break
```

```python
# Custom training loop for FixMatch
epochs = 20
steps_per_epoch = min(len(train_generator), len(unlabeled_weak_generator))

for epoch in range(epochs):
    print(f'\nEpoch {epoch+1}/{epochs}')
    progbar = tf.keras.utils.Progbar(steps_per_epoch)
    
    for step in range(steps_per_epoch):
        # Labeled data batch
        labeled_x, labeled_y = next(train_generator)
        
        # Unlabeled data batches (weak and strong)
        try:
            weak_x = next(unlabeled_weak_generator)
            strong_x = next(unlabeled_strong_generator)
        except StopIteration:
            unlabeled_weak_generator.reset()
            unlabeled_strong_generator.reset()
            weak_x = next(unlabeled_weak_generator)
            strong_x = next(unlabeled_strong_generator)
        
        # Compute losses using GradientTape
        with tf.GradientTape() as tape:
            # Pseudo-labels from weak augmentations
            pseudo_probs = model(weak_x, training=False)
            pseudo_labels = tf.argmax(pseudo_probs, axis=1)
            confidence = tf.reduce_max(pseudo_probs, axis=1)
            mask = confidence >= confidence_threshold
            mask_float = tf.cast(mask, pseudo_probs.dtype)  # Now matches prediction dtype
            num_confident = tf.reduce_sum(mask_float)
            
            # Supervised loss
            labeled_preds = model(labeled_x, training=True)
            labeled_loss = tf.keras.losses.categorical_crossentropy(labeled_y, labeled_preds)
            labeled_loss = tf.reduce_mean(labeled_loss)
            
            # Unsupervised loss
            strong_preds = model(strong_x, training=True)
            unsupervised_loss = tf.keras.losses.sparse_categorical_crossentropy(pseudo_labels, strong_preds)
            unsupervised_loss = tf.reduce_sum(unsupervised_loss * mask_float) / (num_confident + 1e-7)
            
            # Total loss
            total_loss = labeled_loss + lambda_u * unsupervised_loss
        
        # Update model weights
        gradients = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        
        progbar.update(step + 1, [('total_loss', total_loss), ('labeled_loss', labeled_loss), ('unsupervised_loss', unsupervised_loss)])
    
    # Validation
    test_loss, test_acc = model.evaluate(test_generator, steps=test_generator.samples // batch_size, verbose=0)
    print(f'\nValidation loss: {test_loss:.4f}, Validation accuracy: {test_acc:.4f}')
```

```python
# Save the model
model.save('FixM_Models/baseline_fixmatch_model.h5')
```

```python
# Evaluate final model
test_loss, test_acc = model.evaluate(test_generator, steps=test_generator.samples // batch_size)
print(f'Final Test Accuracy: {test_acc:.4f}')
```

```python
# Define weak and strong augmentations
weak_aug = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

strong_aug = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2]
)

# Create generators
labeled_train = weak_aug.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
)

unlabeled_weak = weak_aug.flow_from_directory(
    directory=os.path.dirname(unlabeled_dir),
    classes=[os.path.basename(unlabeled_dir)],
    target_size=img_size,
    batch_size=batch_size,
    class_mode=None,
    shuffle=True,
    seed=42,
)

unlabeled_strong = strong_aug.flow_from_directory(
    directory=os.path.dirname(unlabeled_dir),
    classes=[os.path.basename(unlabeled_dir)],
    target_size=img_size,
    batch_size=batch_size,
    class_mode=None,
    shuffle=True,
    seed=42,
)

# Custom training loop
optimizer = tf.keras.optimizers.Adam(1e-5)
lambda_u = 1.0
confidence_threshold = 0.95

for epoch in range(10):
    print(f'\nEpoch {epoch+1}/10')
    progbar = tf.keras.utils.Progbar(len(labeled_train))
    
    for step in range(len(labeled_train)):
        # Get labeled batch
        (x_labeled, y_labeled) = labeled_train[step]
        
        # Get unlabeled batches
        x_weak = unlabeled_weak[step]
        x_strong = unlabeled_strong[step]
        
        with tf.GradientTape() as tape:
            # Pseudo-label with weak augmentation
            pseudo_probs = model(x_weak, training=False)
            pseudo_labels = tf.argmax(pseudo_probs, axis=1)
            confidences = tf.reduce_max(pseudo_probs, axis=1)
            
            # Create mask for confident predictions
            mask = confidences >= confidence_threshold
            mask_float = tf.cast(mask, pseudo_probs.dtype)
            
            # Calculate supervised loss
            y_pred = model(x_labeled, training=True)
            loss_labeled = tf.reduce_mean(
                tf.keras.losses.categorical_crossentropy(y_labeled, y_pred)
            )
            
            # Calculate unsupervised loss
            y_pred_strong = model(x_strong, training=True)
            loss_unlabeled = tf.reduce_sum(
                tf.keras.losses.sparse_categorical_crossentropy(pseudo_labels, y_pred_strong) * mask_float
            ) / (tf.reduce_sum(mask_float) + 1e-7)
            
            # Total loss
            total_loss = loss_labeled + lambda_u * loss_unlabeled
        
        # Update weights
        gradients = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        
        progbar.update(step+1, [('total_loss', total_loss), 
                                ('supervised', loss_labeled), 
                                ('unsupervised', loss_unlabeled)])
    
    # Validation
    val_loss, val_acc = model.evaluate(test_generator, verbose=0)
    print(f'\nValidation loss: {val_loss:.4f} - Accuracy: {val_acc:.4f}')

model.save('FixM_Models/mobilenet_fixmatch.h5')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to display predictions
def display_predictions(model, test_generator, num_examples=5):
    # Get one batch of test data
    x_test, y_true = test_generator[1]  # Use __getitem__ to get the batch
    y_true = np.argmax(y_true, axis=1)  # Convert one-hot to class indices
    
    # Predict classes
    y_pred = np.argmax(model.predict(x_test), axis=1)
    
    # Map class indices to class names
    class_names = list(test_generator.class_indices.keys())
    
    # Display images with predictions
    plt.figure(figsize=(15, 5))
    for i in range(num_examples):
        plt.subplot(1, num_examples, i + 1)
        plt.imshow(x_test[i])  # Display the image
        plt.axis('off')
        
        true_label = class_names[y_true[i]]
        pred_label = class_names[y_pred[i]]
        is_correct = true_label == pred_label
        
        # Add title with true/predicted labels and correctness
        plt.title(f"True: {true_label}\nPred: {pred_label}\nCorrect: {is_correct}", 
                  color='green' if is_correct else 'red')
    
    plt.tight_layout()
    plt.show()

# Call the function to display predictions
display_predictions(model, test_generator, num_examples=10)
```
